## TF-IDF(Term Frequency - Inverse Document Frequency)を求める

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df = pd.DataFrame({'id': ['A', 'B'],
                   'text': [ 'プログラミング 勉強 プログラマー 独学 サイト 転職 スキル スクール エンジニア 本 言語 必要 おすすめ 時間 人 Web 方法 経験 学習 情報 パソコン 未 性 入門 動画 法 仕事 初心者 ノート 資格 紹介 やり方 面接', 'プログラミング スクール Web チュートリアル 独学 人 転職 おすすめ エンジニア 学習 勉強 副業 制作 体験 CodeCamp 機能 方法 基礎 プロダクト ポテパンキャンプ 経験 作業 メンター 仕事 初心者 TechAcademy']})


# TF-IDFの計算
tfidf_vectorizer = TfidfVectorizer(use_idf=True,lowercase=False)

# 文章内の全単語のTfidf値を取得
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

# index 順の単語リスト
terms = tfidf_vectorizer.get_feature_names()

# 単語毎のtfidf値配列：TF-IDF 行列 (numpy の ndarray 形式で取得される)
# 1つ目の文書に対する、各単語のベクトル値
# 2つ目の文書に対する、各単語のベクトル値
# ・・・
# が取得できる（文書の数 * 全単語数）の配列になる。（toarray()で密行列に変換）
tfidfs = tfidf_matrix.toarray()
```

```python
print(terms)
```

```python
print(tfidfs)
```
