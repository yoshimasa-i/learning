```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df = pd.DataFrame({'id': ['A', 'B'],
                   'text': [ 'シューズ ウェア バッグ アクセサリ ジュニア スケート LAKAI ボード ラカイ メンズ インナー サングラス トレーニング マシン スイムウェア 用品 スケボー ポート NEWPORT 在庫 切れ 通販 スポタカ ブランド アウトドア ブーツ サンダル デッキ トラック ウィール ベアリング', 'アイテム シューズ スケート SHOES LAKAI デッキ ラカイ 一覧 その他 ボード WHEELS KIDS 指 スケ 定価 FOOTWEAR モデル SKATEBOARD 組み立て DECK ブランド OTHER レディース ブラック LIMITED トリック 用 CRUISER ロング 別 SKATE ハードウェア BOARD ACCESSORIES アクセサリー']})


# TF-IDFの計算
tfidf_vectorizer = TfidfVectorizer(use_idf=True,lowercase=False)

# 文章内の全単語のTfidf値を取得
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

# index 順の単語リスト
terms = tfidf_vectorizer.get_feature_names()

# 単語毎のtfidf値配列：TF-IDF 行列 (numpy の ndarray 形式で取得される)
# 1つ目の文書に対する、各単語のベクトル値
# 2つ目の文書に対する、各単語のベクトル値
# ・・・
# が取得できる（文書の数 * 全単語数）の配列になる。（toarray()で密行列に変換）
tfidfs = tfidf_matrix.toarray()
```
